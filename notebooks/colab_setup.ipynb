```python
####
#   COPY THIS CELL TO SETUP THE ENVIRONMENT IN COLAB!!!
###

from google.colab import drive
drive.mount('/content/drive')

import os
import sys
import shutil
import importlib

repo_url = "https://github.com/ochiga07/ml_final_project.git"
repo_path = "/content/ml_final_project"
drive_repo = "/content/drive/MyDrive/ml_final_project"

if not os.path.exists(repo_path):
    !git clone {repo_url} {repo_path}
else:
    %cd {repo_path}
    !git pull

if os.path.exists(drive_repo):
    for item in os.listdir(repo_path):
        src = os.path.join(repo_path, item)
        dst = os.path.join(drive_repo, item)

        if item == "data":
            continue

        if os.path.isdir(src):
            if os.path.exists(dst):
                shutil.rmtree(dst)
            shutil.copytree(src, dst)
        else:
            shutil.copy2(src, dst)
else:
    shutil.copytree(repo_path, drive_repo)

if drive_repo not in sys.path:
    sys.path.append(drive_repo)

import feature_pipeline
importlib.reload(feature_pipeline)

from feature_pipeline import load_raw_data, run_feature_pipeline, split_and_save


train, test, features, stores = load_raw_data(
    path='/content/drive/MyDrive/ml_final_project/data/'
)

full_df = run_feature_pipeline(train, test, features, stores)
processed_train, processed_test = split_and_save(full_df)
```
